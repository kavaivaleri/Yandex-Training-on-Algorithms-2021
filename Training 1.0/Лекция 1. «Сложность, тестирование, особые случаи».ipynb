### Задача 1
Найти самую часто встречаемую букву в string

#### Решение 1
Время: $O(N^2)$
Память: $O(N)$

```python
s = input()
ans = ''
anscnt = 0

for i in range(len(s)):
    nowcnt = 0
    for j in range(len(s)):
        if s[i] == s[j]:
            nowcnt += 1
    if nowcnt > anscnt:
        ans = s[i]
        anscnt = nowcnt
print(ans)
```

#### Решение 2
Время: $O(Nk)$
Память: $O(N+k)=O(N)$

Переберем все символа, встречающиеся в строке, а затем переберем все позиции. И в случае совпадения прибавим к счетчику единицу. Найдём максимальное значение счётчика.

Есть функция set(), которая из многих объектов делает множество значений, которые туда входят.

Теперь решение занимает $O(Nk)$

```python
s = input()
ans = ''
anscnt = 0

for i in set(s):
    nowcnt = 0
    for j in range(len(s)):
        if s[i] == s[j]:
            nowcnt += 1
    if nowcnt > anscnt:
        ans = s[i]
        anscnt = nowcnt
print(ans)
```
